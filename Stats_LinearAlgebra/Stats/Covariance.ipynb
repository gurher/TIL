### Variance and Covariance

Population Variance
$$\sigma^2_{x} = \frac{1}{N}\sum^n_{i=1}(x_{i}-\overline{x})^2 $$


Covariance
$$\sigma^{2}_{x,y} = \frac{1}{N}\sum_{i=1}^n(x_i - \overline{x}_n)(y_i - \overline{y}_n)^\top$$


The diagonal entries of the covariance matrix are the variances and the other entries are the covariances.

$$\begin{pmatrix}
\sigma^2_x & \sigma_{xy} & \sigma_{xz} \\
\sigma_{yx} & \sigma^2_y & \sigma_{yz} \\
\sigma_{zx} & \sigma_{zy} & \sigma^2_z 
\end{pmatrix}$$




```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

x = np.random.normal(0, 2, 5000)
y = np.random.normal(2, 3, 5000)
z = np.random.normal(1, 1.5, 5000)
data = np.vstack((x, y, z)).T

# Plot
df=pd.DataFrame(data)
df.columns=["x","y","z"]

fig = plt.figure(figsize = (15, 15) )
ax = plt.axes(projection='3d')

zdata = df['z']
xdata = df['x']
ydata = df['y']
ax.scatter3D(xdata, ydata, zdata, c=(xdata+ydata+zdata))  #, cmap='Greens');
ax.view_init(0, 45)

```

```python
df.cov()
```

```python
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12, 8)

# Normal distributed x and y vector with mean 0 and standard deviation 1
x = np.random.normal(10, 6, 5000)
y = np.random.normal(2, 2, 5000)
data = np.vstack((x, y)).T

plt.scatter(data[:, 0], data[:, 1])
plt.title('Generated Data')
plt.axis('equal');
```

```python
df = pd.DataFrame(data, columns=['x','y'])
df.cov()
```


This case would mean that x and y are independent (or uncorrelated) and the covariance matrix 
C is
$$\begin{pmatrix}
\sigma^2_x & 0 \\
0 & \sigma^2_y
\end{pmatrix}$$
